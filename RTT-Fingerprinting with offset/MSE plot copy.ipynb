```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

data = {
    "Model": ["DNN","DNN+KNN", "Proposed DNN+RF"],
    "MSE": [0.41, 0.19, 0.110],
    "RMSE": [0.64, 0.44, 0.331],
    
   
}

df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10,6))
models = df["Model"]
MSE = df["MSE"]
RMSE = df["RMSE"]

# Set the width of the bars
bar_width = 0.2
# Set the x positions of the bars
r1 = range(len(models))
r2 = [x + bar_width for x in r1]

# Create the bar chart
plt.bar(r1, MSE, width=bar_width, color=["grey"], hatch="/"*5)
plt.bar(r2, RMSE, width=bar_width, color=["red" ], hatch="x"*5)

# Set the labels and title
plt.xticks([r + bar_width/2 for r in range(len(models))], models)
plt.ylabel("Mean Squared Error in m")
plt.title("Model Performance Comparision")

# Add a legend
plt.legend(["MSE", "RMSE"])

# Show the plot
plt.show()
```

```python
data1 = {
    "Model": ["DNN","DNN+KNN", "Proposed DNN+RF"],
    "MAPE": [20.60, 9.56, 8.19],
    "R2 Score": [88.87, 94.20, 96.95],
    
   
}

df1 = pd.DataFrame(data1)
```

```python
plt.figure(figsize=(10,6))
models = df1["Model"]
MAPE = df1["MAPE"]
R2_Score = df1["R2 Score"]

# Set the width of the bars
bar_width = 0.2
# Set the x positions of the bars
r1 = range(len(models))
r2 = [x + bar_width for x in r1]

# Create the bar chart
plt.bar(r1, MAPE, width=bar_width, color=["grey"], hatch="/"*5)
plt.bar(r2, R2_Score, width=bar_width, color=["red" ], hatch="x"*5)

# Set the labels and title
plt.xticks([r + bar_width/2 for r in range(len(models))], models)
plt.ylabel("Mean Squared Error in m")
plt.title("Model Performance Comparision")

# Add a legend
plt.legend(["MAPE", "R2 Score"])

# Show the plot
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,6))

models = df["Model"]
MSE = df["MSE"]
RMSE = df["RMSE"]

# Set the width of the bars
bar_width = 0.2
# Set the x positions of the bars
r1 = range(len(models))
r2 = [x + bar_width for x in r1]

# Create the bar chart for the first subplot
ax1.bar(r1, MSE, width=bar_width, color=["grey"], hatch="/"*5)
ax1.bar(r2, RMSE, width=bar_width, color=["red" ], hatch="x"*5)

# Set the labels and title for the first subplot
ax1.set_xticks([r + bar_width/2 for r in range(len(models))])
ax1.set_xticklabels(models)
ax1.set_ylabel("Mean Squared Error in m")
ax1.set_title("Model Performance Comparision")
ax1.legend(["MSE", "RMSE"])

models = df1["Model"]
MAPE = df1["MAPE"]
R2_Score = df1["R2 Score"]

# Set the width of the bars
bar_width = 0.2
# Set the x positions of the bars
r1 = range(len(models))
r2 = [x + bar_width for x in r1]

# Create the bar chart for the second subplot
ax2.bar(r1, MAPE, width=bar_width, color=["grey"], hatch="/"*5)
ax2.bar(r2, R2_Score, width=bar_width, color=["red" ], hatch="x"*5)

# Set the labels and title for the second subplot
ax2.set_xticks([r + bar_width/2 for r in range(len(models))])
ax2.set_xticklabels(models)
ax2.set_ylabel("R2 Score")
ax2.set_title("Model Performance Comparision")
ax2.legend(["MAPE", "R2 Score"])

# Adjust the spacing between the subplots
fig.subplots_adjust(wspace=0.4)

# Show the plot
plt.show()
```

The given values represent the performance metrics (MAPE and R2 Score) of three different models: DNN, DNN+KNN, and Proposed DNN+RF.

MAPE (Mean Absolute Percentage Error) is a measure of the average deviation of the predicted values from the actual values as a percentage of the actual values. A lower MAPE value indicates better accuracy of the model.

R2 Score is a statistical measure that represents the proportion of the variance in the dependent variable (i.e., target variable) that is predictable from the independent variables (i.e., input features). It ranges from 0 to 1, where 1 represents perfect prediction and 0 represents no correlation. A higher R2 Score value indicates better performance of the model.

From the given values, we can observe that the Proposed DNN+RF model outperforms the other two models in both MAPE and R2 Score. The MAPE value of Proposed DNN+RF is the lowest among all the models, indicating the highest accuracy in predicting the target values. The R2 Score of Proposed DNN+RF is also the highest among all the models, indicating the highest correlation between the input features and target values.

In contrast, the DNN model has the highest MAPE value and the lowest R2 Score, indicating the lowest accuracy and correlation, respectively. The DNN+KNN model shows better performance than the DNN model, but it still has a higher MAPE value and a lower R2 Score than the Proposed DNN+RF model.

Therefore, based on the given performance metrics, the Proposed DNN+RF model is the best-performing model among the three models.

```python
import matplotlib.pyplot as plt
import numpy as np

# define the given data
model = ["DNN only","DNN+KNN", "Proposed DNN+RF"]
mape = [20.60, 9.56, 8.19]
r2_score = [0.8887, 0.9420, 0.9695]
mse = [0.41, 0.19, 0.110]
rmse = [0.64, 0.44, 0.331]

# plot MAPE and R2 score for each model
x = np.arange(len(model))
width = 0.4
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, mape, width, label='MAPE')
rects2 = ax.bar(x + width/2, r2_score, width, label='R2 Score')
ax.set_xticks(x)
ax.set_xticklabels(model)
ax.legend()
plt.ylabel("Percentage")
plt.title("Model Performance Comparision")
plt.show()

# plot MSE and RMSE for each model
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, mse, width, label='MSE')
rects2 = ax.bar(x + width/2, rmse, width, label='RMSE')
ax.set_xticks(x)
ax.set_xticklabels(model)
ax.legend()
plt.ylabel(" Error in m")
plt.title("Model Performance Comparision")
plt.show()

# print the best model based on MAPE, R2 Score, and RMSE
best_mape = model[np.argmin(mape)]
best_r2_score = model[np.argmax(r2_score)]
best_rmse = model[np.argmin(rmse)]
print(f"The best model based on MAPE is {best_mape}")
print(f"The best model based on R2 Score is {best_r2_score}")
print(f"The best model based on RMSE is {best_rmse}")
```

Both formats for presenting the R2 score are accurate.

The R2 score is a metric that represents the proportion of the variance in the dependent variable (i.e., the target variable) that can be explained by the independent variables (i.e., the input features) in a regression model. The value of R2 ranges between 0 and 1, where 0 indicates that the model cannot explain any of the variance in the target variable, and 1 indicates that the model can perfectly explain all of the variance.

So, in the given example, both [88.87, 94.20, 96.95] and [0.8887, 0.9420, 0.9695] represent the same R2 scores for three different models, and both are accurate representations of the R2 score values.

However, presenting R2 scores in decimal format (i.e., [0.8887, 0.9420, 0.9695]) can make it easier to interpret and compare the scores since the range is between 0 and 1. Therefore, it is generally more common to present R2 scores in decimal format.
