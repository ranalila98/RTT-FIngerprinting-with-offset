```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

data = {
    "Model": ["DNN", "DNN+KNN", "DNN+RF"],
    "Without Calibrated data": [0.49, 0.20, 0.162],
    "With Calibrated data": [0.41, 0.19, 0.110],
    
    "Has features": [True, True, True]
}

df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(8,6))
models = df["Model"]
no_feature_values = df["Without Calibrated data"]
feature_values = df["With Calibrated data"]
has_features = df["Has features"]

# Set the width of the bars
bar_width = 0.3
# Set the x positions of the bars
r1 = range(len(models))
r2 = [x + bar_width for x in r1]

# Create the bar chart
plt.bar(r1, no_feature_values, width=bar_width, color=["grey" if not f else "blue" for f in has_features], hatch="/"*5)
plt.bar(r2, feature_values, width=bar_width, color=["red" if f else "grey" for f in has_features], hatch="x"*5)

# Set the labels and title
plt.xticks([r + bar_width/2 for r in range(len(models))], models)
plt.ylabel("Mean Squared Error in m")
#plt.title("Model comparision for test points in Reference Point")

# Add a legend
plt.legend(["Raw Data", "Calibrated Data"])

# Show the plot
plt.show()
```
