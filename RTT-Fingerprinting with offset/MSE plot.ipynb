```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

data = {
    "Model": ["KNN", "RF", "DNN", "DNN+KNN", "Proposed DNN+RF"],
    "No features Extraction": [1.13, 1.05, 0.44, 0.33, 0.099],
    "Features Extraction": [0.9, 0.704, 0.33, 0.3, 0.063],
    
    "Has features": [True, True, True, True, True]
}

df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10,6))
models = df["Model"]
no_feature_values = df["No features Extraction"]
feature_values = df["Features Extraction"]
has_features = df["Has features"]

# Set the width of the bars
bar_width = 0.4
# Set the x positions of the bars
r1 = range(len(models))
r2 = [x + bar_width for x in r1]

# Create the bar chart
plt.bar(r1, no_feature_values, width=bar_width, color=["grey" if not f else "blue" for f in has_features], hatch="/"*5)
plt.bar(r2, feature_values, width=bar_width, color=["red" if f else "grey" for f in has_features], hatch="x"*5)

# Set the labels and title
plt.xticks([r + bar_width/2 for r in range(len(models))], models)
plt.ylabel("Mean Squared Error in m")
plt.title("Model Performance Comparision")

# Add a legend
plt.legend(["No features Extraction", "Features extraction"])

# Show the plot
plt.show()
```
