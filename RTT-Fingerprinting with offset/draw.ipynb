```python
import matplotlib.pyplot as plt

# Set up the room coordinates
x = [0, 7, 7, 0, 0]
y = [0, 0, 9, 9, 0]

# Create the figure and axes objects
fig, ax = plt.subplots()

# Plot the room as a polygon
ax.plot(x, y, color='black')

# Set the x and y limits to leave a margin of 1 unit around the edges
ax.set_xlim(-1, 8)
ax.set_ylim(-1, 10)

# Generate grid of coordinates with 1 unit gap
x_grid = range(1, 7, 1)
y_grid = range(1, 9, 1)

# Plot points at each grid location
for i in x_grid:
    for j in y_grid:
        ax.scatter(i, j, marker='o', color='black', s=20)

# Highlight specific points with a red star symbol
ax.scatter(7, 0, marker='*', color='red', s=150)
ax.scatter(0, 9, marker='*', color='red', s=150)
ax.scatter(7, 9, marker='*', color='red', s=150)

# Show the plot
plt.show()


```

```python
import matplotlib.pyplot as plt

# Set up the room coordinates
x = [0, 7, 7, 0, 0]
y = [0, 0, 9, 9, 0]

# Define a list of excluded points
excluded_points = [(1,6),(1,5), (1,4),(2,6),(2,5), (2,4), (4,6), (4,5), (4,7)]

# Create the figure and axes objects
fig, ax = plt.subplots()

# Plot the room as a polygon
ax.plot(x, y, color='black')

# Set the x and y limits to match the room coordinates
ax.set_xlim(0, 7)
ax.set_ylim(0, 9)

# Generate grid of coordinates with 1 unit gap
x_grid = range(1, 7, 1)
y_grid = range(1, 9, 1)

# Plot points at each grid location, excluding the excluded points
for i in x_grid:
    for j in y_grid:
        if (i,j) not in excluded_points:
            ax.scatter(i, j, marker='o', color='black', s=30)

# Highlight specific points with a red star symbol
ax.scatter(7, 0, marker='*', color='red', s=200)
ax.scatter(0, 9, marker='*', color='red', s=200)
ax.scatter(7, 9, marker='*', color='red', s=200)



# Set the x and y limits to leave a margin of 1 unit around the edges
ax.set_xlim(-1, 8)
ax.set_ylim(-1, 10)
# Show the plot
plt.show()
```
