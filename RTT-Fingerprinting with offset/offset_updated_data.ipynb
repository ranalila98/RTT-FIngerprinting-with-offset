```python
#it is for offset code for random test data
actual_co_ordinate = (1,2.5)

# Set the path to the folder containing the dataset
data_folder = r'C:\Users\LILA\Desktop\RTT-FIngerprinting-with-offset\RTT-Fingerprinting with offset\Dataset\Random_test_data\raw'

# Loop through each CSV file in the folder and read it into a DataFrame
for file_name in os.listdir(data_folder):
     print(os.listdir(data_folder))
     if file_name.endswith('.csv'): # Check if the file is a CSV file
        file_parts = file_name[:-4].split('_') # Remove the '.csv' extension and split the filename
        if len(file_parts) < 2: # Check if there are at least two parts in the filename
            continue # Skip the file if there are not enough parts
        x, y = float(file_parts[0]), float(file_parts[1]) # Convert the two parts to floats
        if (x, y) == actual_co_ordinate: # Check if the coordinates match
            file_path = os.path.join(data_folder, file_name)
            df = pd.read_csv(file_path)
            break

df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# Drop rows with NaN values
df.dropna(inplace=True)
#print(df.head())  # Print the first 5 rows of the DataFrame
df
```

```python
df['True_range1']=np.sqrt((df.x-df.x1)**2+(df.y-df.y1)**2)
df['True_range2']=np.sqrt((df.x-df.x2)**2+(df.y-df.y2)**2)
df['True_range3']=np.sqrt((df.x-df.x3)**2+(df.y-df.y3)**2)
df
```

```python
offset1=np.mean(df.AP1RTT-df.True_range1)
offset2=np.mean(df.AP2RTT-df.True_range2)
offset3=np.mean(df.AP3RTT-df.True_range3)
print(offset1)
print(offset2)
print(offset3)
```

```python
df['AP1RTT']= df['AP1RTT']-offset1
df['AP2RTT']= df['AP2RTT']-offset2
df['AP3RTT']= df['AP3RTT']-offset3
df
```

```python
# # Specify the file name and path where you want to save the file
file_path = 'Offset_Test_random'+str(actual_co_ordinate)+'.csv'


# Save the DataFrame to a CSV file using the to_csv() method
df.to_csv(file_path, index=False)

print(f'File saved to {file_path}!')
```
