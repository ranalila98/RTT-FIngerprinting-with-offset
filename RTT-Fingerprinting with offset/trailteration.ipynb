```python
import pandas as pd

# Define the function to calculate phone location
def trackPhone(x1, y1, r1, x2, y2, r2, x3, y3, r3):
    A = 2*x2 - 2*x1
    B = 2*y2 - 2*y1
    C = r1**2 - r2**2 - x1**2 + x2**2 - y1**2 + y2**2
    D = 2*x3 - 2*x2
    E = 2*y3 - 2*y2
    F = r2**2 - r3**2 - x2**2 + x3**2 - y2**2 + y3**2
    x = (C*E - F*B) / (E*A - B*D)
    y = (C*D - A*F) / (B*D - A*E)
    return x, y

# Define the access point coordinates
AP1 = (7, 9)
AP2 = (7, 0)
AP3 = (0, 9)

# Define the distances from the cell phone to each access point
distances = dfff[['AP1RTT', 'AP2RTT', 'AP3RTT']]

# Loop through the rows of the DataFrame and calculate the phone location for each row
locations = []
for index, row in distances.iterrows():
    d1, d2, d3 = row['AP1RTT'], row['AP2RTT'], row['AP3RTT']
    x1, y1, r1 = AP1[0], AP1[1], d1
    x2, y2, r2 = AP2[0], AP2[1], d2
    x3, y3, r3 = AP3[0], AP3[1], d3
    x, y = trackPhone(x1, y1, r1, x2, y2, r2, x3, y3, r3)
    locations.append((x, y))

# Add the phone locations to the DataFrame
dfff['predicted_x'] = [location[0] for location in locations]
dfff['predicted_y'] = [location[1] for location in locations]

# Save the updated DataFrame to a new file
# df.to_csv('your_updated_dataframe.csv', index=False)
dfff
```

```python
import pandas as pd
from sklearn.metrics import mean_squared_error
import numpy as np

# extract the true and predicted values
true_x = dfff['x']
true_y = dfff['y']
pred_x = dfff['predicted_x']
pred_y = dfff['predicted_y']

# calculate MSE and RMSE for x and y coordinates
mse_x = mean_squared_error(true_x, pred_x)
rmse_x = np.sqrt(mse_x)
mse_y = mean_squared_error(true_y, pred_y)
rmse_y = np.sqrt(mse_y)


mse= (mse_x + mse_y) / 2
rmse=(rmse_x + rmse_y) / 2
# print the results
print(f"MSE : {mse:.4f}")
print(f"RMSE : {rmse:.4f}")
```
